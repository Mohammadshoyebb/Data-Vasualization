```python
import pandas as pd
import numpy as np
import seaborn as sns
tips=sns.load_dataset('tips')
tips.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(tips['total_bill'],tips['tip'])
plt.title("This is Scatterplot")
plt.xlabel("Total Bill")
plt.ylabel("Tips")
```

```python
plt.scatter(tips['day'],tips['tip'])
plt.title("This is Scatterplot")
plt.xlabel("day")
plt.ylabel("Tips")
plt.show()
```

```python
import plotly.express
graph=plotly.express.scatter(tips,x="total_bill",y='tip',color='smoker')
graph.show()
```
